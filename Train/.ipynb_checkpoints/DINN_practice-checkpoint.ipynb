```python
# 학습용 및 예측 경계 설정
Ds=168
```

```python
%%time

PATH = f'model/covid_real_data_cumulative_cases_{eng[city]}' 

class DINN(nn.Module):
    def __init__(self, t, S_data, I_data1, I_data2, D_data1, R_data1, D_data2, R_data2, alpha1, device): #[t,S,I1,I2,D,R,a1]
        super(DINN, self).__init__()
        self.device = device
        self.N = pop_N[city] #population size
        self.t = torch.tensor(t, requires_grad=True, device = self.device)
        self.t_float = self.t.float()
        self.t_batch = torch.reshape(self.t_float, (len(self.t),1)) #reshape for batch 
        self.S = torch.tensor(S_data, device = self.device)
        self.I1 = torch.tensor(I_data1, device = self.device)
        self.I2 = torch.tensor(I_data2, device = self.device)
        self.D1 = torch.tensor(D_data1, device = self.device)
        self.R1 = torch.tensor(R_data1, device = self.device)
        self.D2 = torch.tensor(D_data2, device = self.device)
        self.R2 = torch.tensor(R_data2, device = self.device)
        self.losses = []
        self.save = 3 #which file to save to

        self.alpha1_tilda = torch.tensor(alpha1, device = self.device)
        self.beta1_tilda = torch.tensor(0.0714, device = self.device) 
        self.gamma1_tilda = torch.tensor(0.008, device = self.device) 
        
        self.alpha2_tilda = torch.nn.Parameter(torch.rand(1, requires_grad=True, device = self.device))
        self.beta2_tilda = torch.nn.Parameter(torch.rand(1, requires_grad=True, device = self.device))
        self.gamma2_tilda = torch.nn.Parameter(torch.rand(1, requires_grad=True, device = self.device))
        
        # find values for normalization
        self.S_max = max(self.S)
        self.I1_max = max(self.I1)
        self.I2_max = max(self.I2)
        self.D1_max = max(self.D1)
        self.R1_max = max(self.R1)
        self.D2_max = max(self.D2)
        self.R2_max = max(self.R2)
        self.S_min = min(self.S)
        self.I1_min = min(self.I1)
        self.I2_min = min(self.I2)
        self.D1_min = min(self.D1)
        self.R1_min = min(self.R1)
        self.D2_min = min(self.D2)
        self.R2_min = min(self.R2)
        
        # normalize
        self.S_hat = (self.S - self.S_min) / (self.S_max - self.S_min)
        self.I1_hat = (self.I1 - self.I1_min) / (self.I1_max - self.I1_min)
        self.I2_hat = (self.I2 - self.I2_min) / (self.I2_max - self.I2_min)
        self.D1_hat = (self.D1 - self.D1_min) / (self.D1_max - self.D1_min)
        self.R1_hat = (self.R1 - self.R1_min) / (self.R1_max - self.R1_min)
        self.D2_hat = (self.D2 - self.D2_min) / (self.D2_max - self.D2_min)
        self.R2_hat = (self.R2 - self.R2_min) / (self.R2_max - self.R2_min)

        # matrices (x4 for S,I1,I2,D,R,a1,a2) for the gradients
        self.m1 = torch.zeros((len(self.t), 7), device = self.device); self.m1[:, 0] = 1
        self.m2 = torch.zeros((len(self.t), 7), device = self.device); self.m2[:, 1] = 1
        self.m3 = torch.zeros((len(self.t), 7), device = self.device); self.m3[:, 2] = 1
        self.m4 = torch.zeros((len(self.t), 7), device = self.device); self.m4[:, 3] = 1
        self.m5 = torch.zeros((len(self.t), 7), device = self.device); self.m5[:, 4] = 1
        self.m6 = torch.zeros((len(self.t), 7), device = self.device); self.m6[:, 5] = 1
        self.m7 = torch.zeros((len(self.t), 7), device = self.device); self.m7[:, 6] = 1
        
        # NN
        self.net_sidr = self.Net_sidr(self.device)
        self.params = list(self.net_sidr.parameters())
        self.params.extend(list([self.alpha1_tilda, self.beta1_tilda, self.gamma1_tilda, 
                                 self.alpha2_tilda, self.beta2_tilda, self.gamma2_tilda]))

    # force parameters to be in a range
    @property
    def alpha1(self):
        return torch.relu(self.alpha1_tilda)

    @property
    def beta1(self):
        return torch.relu(self.beta1_tilda)
    
    @property
    def gamma1(self):
        return torch.relu(self.gamma1_tilda)
    
    @property
    def alpha2(self):
        return torch.relu(self.alpha2_tilda)
      
    @property
    def beta2(self):
        return torch.relu(self.beta2_tilda)
    
    @property
    def gamma2(self):
        return torch.relu(self.gamma2_tilda)


    #nets
    class Net_sidr(nn.Module): # input = [t]
        def __init__(self, device):
            super(DINN.Net_sidr, self).__init__()
            self.device = device
            self.fc1=nn.Linear(1, 20).to(self.device)
            self.fc2=nn.Linear(20, 20).to(self.device)
            self.fc3=nn.Linear(20, 20).to(self.device)
            self.fc4=nn.Linear(20, 20).to(self.device)
            self.fc5=nn.Linear(20, 20).to(self.device)
            self.fc6=nn.Linear(20, 20).to(self.device)
            self.fc7=nn.Linear(20, 20).to(self.device)
            self.fc8=nn.Linear(20, 20).to(self.device)
            self.fc9=nn.Linear(20, 20).to(self.device)
            self.fc10=nn.Linear(20, 20).to(self.device)
            self.fc11=nn.Linear(20, 20).to(self.device)
            self.fc12=nn.Linear(20, 20).to(self.device)
            self.fc13=nn.Linear(20, 20).to(self.device)
            self.fc14=nn.Linear(20, 20).to(self.device)
            self.fc15=nn.Linear(20, 20).to(self.device)
            self.fc16=nn.Linear(20, 20).to(self.device)
            self.out=nn.Linear(20, 7).to(self.device) #outputs S, I1, I2, D1, D2, R1, R2

        def forward(self, t_batch):
            sidr=F.gelu(self.fc1(t_batch))
            sidr=F.gelu(self.fc2(sidr))
            sidr=F.gelu(self.fc3(sidr))
            sidr=F.gelu(self.fc4(sidr))
            sidr=F.gelu(self.fc5(sidr))
            sidr=F.gelu(self.fc6(sidr))
            sidr=F.gelu(self.fc7(sidr))
            sidr=F.gelu(self.fc8(sidr))
            sidr=F.gelu(self.fc9(sidr))
            sidr=F.gelu(self.fc10(sidr))
            sidr=F.gelu(self.fc11(sidr))
            sidr=F.gelu(self.fc12(sidr))
            sidr=F.gelu(self.fc13(sidr))
            sidr=F.gelu(self.fc14(sidr))
            sidr=F.gelu(self.fc15(sidr))
            sidr=F.gelu(self.fc16(sidr))
            sidr=self.out(sidr)
            return sidr
            
    def net_f(self, t_batch):
        sidr_hat = self.net_sidr(t_batch)
        S_hat, I1_hat, I2_hat, D1_hat, R1_hat, D2_hat, R2_hat = sidr_hat[:,0], sidr_hat[:,1], sidr_hat[:,2], sidr_hat[:,3], sidr_hat[:,4], sidr_hat[:,5], sidr_hat[:,6]

        #S_t
        sidr_hat.backward(self.m1, retain_graph=True)
        
        S_hat_t = self.t.grad.clone()
        self.t.grad.zero_()

        #I1_t
        sidr_hat.backward(self.m2, retain_graph=True)
        
        I1_hat_t = self.t.grad.clone()
        self.t.grad.zero_()
        
        #I2_t
        sidr_hat.backward(self.m3, retain_graph=True)
        
        I2_hat_t = self.t.grad.clone()
        self.t.grad.zero_()

        #D1_t
        sidr_hat.backward(self.m4, retain_graph=True)
        
        D1_hat_t = self.t.grad.clone()
        self.t.grad.zero_()

        #R1_t
        sidr_hat.backward(self.m5, retain_graph=True)
        
        R1_hat_t = self.t.grad.clone()
        self.t.grad.zero_()
        
        #D2_t
        sidr_hat.backward(self.m6, retain_graph=True)
        
        D2_hat_t = self.t.grad.clone()
        self.t.grad.zero_()

        #R2_t
        sidr_hat.backward(self.m7, retain_graph=True)
        
        R2_hat_t = self.t.grad.clone()
        self.t.grad.zero_()

        #unnormalize
        S = self.S_min + (self.S_max - self.S_min) * S_hat
        I1 = self.I1_min + (self.I1_max - self.I1_min) * I1_hat
        I2 = self.I2_min + (self.I2_max - self.I2_min) * I2_hat
        D1 = self.D1_min + (self.D1_max - self.D1_min) * D1_hat      
        R1 = self.R1_min + (self.R1_max - self.R1_min) * R1_hat
        D2 = self.D2_min + (self.D2_max - self.D2_min) * D2_hat      
        R2 = self.R2_min + (self.R2_max - self.R2_min) * R2_hat

        f1_hat = S_hat_t - (-(self.alpha1 / self.N) * S * I1 - (self.alpha2 / self.N) * S * I2)  / (self.S_max - self.S_min)
        f2_hat = I1_hat_t - ((self.alpha1 / self.N) * S * I1 - self.beta1 * I1 - self.gamma1 * I1 ) / (self.I1_max - self.I1_min)
        f3_hat = I2_hat_t - ((self.alpha2 / self.N) * S * I2 - self.beta2 * I2 - self.gamma2 * I2 ) / (self.I2_max - self.I2_min)
        f4_hat = D1_hat_t - (self.gamma1 * I1) / (self.D1_max - self.D1_min)
        f5_hat = R1_hat_t - (self.beta1 * I1) / (self.R1_max - self.R1_min)
        f6_hat = D2_hat_t - (self.gamma2 * I2) / (self.D2_max - self.D2_min)
        f7_hat = R2_hat_t - (self.beta2 * I2) / (self.R2_max - self.R2_min)
        


        return f1_hat, f2_hat, f3_hat, f4_hat, f5_hat, f6_hat, f7_hat, S_hat, I1_hat, I2_hat, D1_hat, R1_hat, D2_hat, R2_hat
    
    
    def load(self):
        # Load checkpoint
        try:
            checkpoint = torch.load(PATH + str(self.save)+'.pt') 
            print('\nloading pre-trained model...')
            self.load_state_dict(checkpoint['model'])
            self.optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
            self.scheduler.load_state_dict(checkpoint['scheduler'])
            epoch = checkpoint['epoch']
         
            self.losses = checkpoint['losses']
         

        except RuntimeError :
            print('changed the architecture, ignore')
            pass
        except FileNotFoundError:
            pass
      
    def train(self, n_epochs):
        #try loading
        self.load()
    
        #train
        print('\nstarting training...\n')
        
        alpha_list = []
        beta_list = []
        gamma_list = []
        loss_min = np.inf
        dinn_iterator = tqdm(range(n_epochs), position = 0)
        for epoch in dinn_iterator:
            #lists to hold the output (maintain only the final epoch)
            S_pred_list = []
            I1_pred_list = []
            I2_pred_list = []
            D1_pred_list = []
            R1_pred_list = []
            D2_pred_list = []
            R2_pred_list = []

            f1, f2, f3, f4, f5, f6, f7, S_pred, I1_pred, I2_pred, D1_pred, R1_pred, D2_pred, R2_pred = self.net_f(self.t_batch)
            self.optimizer.zero_grad()

            S_pred_list.append(self.S_min + (self.S_max - self.S_min) * S_pred)
            I1_pred_list.append(self.I1_min + (self.I1_max - self.I1_min) * I1_pred)
            I2_pred_list.append(self.I2_min + (self.I2_max - self.I2_min) * I2_pred)
            D1_pred_list.append(self.D1_min + (self.D1_max - self.D1_min) * D1_pred)
            R1_pred_list.append(self.R1_min + (self.R1_max - self.R1_min) * R1_pred)
            D2_pred_list.append(self.D2_min + (self.D2_max - self.D2_min) * D2_pred)
            R2_pred_list.append(self.R2_min + (self.R2_max - self.R2_min) * R2_pred)

            loss = (torch.mean(torch.square(self.S_hat[:Ds] - S_pred[:Ds]))+ 
                        torch.mean(torch.square(self.I1_hat[:Ds] - I1_pred[:Ds]))+
                        torch.mean(torch.square(self.I2_hat[:Ds] - I2_pred[:Ds]))+
                        torch.mean(torch.square(self.D1_hat[:Ds] - D1_pred[:Ds]))+
                        torch.mean(torch.square(self.R1_hat[:Ds] - R1_pred[:Ds]))+
                        torch.mean(torch.square(self.D2_hat[:Ds] - D2_pred[:Ds]))+
                        torch.mean(torch.square(self.R2_hat[:Ds] - R2_pred[:Ds]))+
                        torch.mean(torch.square(f1[:Ds]))+
                        torch.mean(torch.square(f2[:Ds]))+
                        torch.mean(torch.square(f3[:Ds]))+
                        torch.mean(torch.square(f4[:Ds]))+
                        torch.mean(torch.square(f5[:Ds]))+
                        torch.mean(torch.square(f6[:Ds]))+
                        torch.mean(torch.square(f7[:Ds]))
                        ) 


            loss.backward()
            self.optimizer.step()
            self.scheduler.step() 

            self.losses.append(loss.item())
            
            alpha_list.append(self.alpha2)
            beta_list.append(self.beta2)
            gamma_list.append(self.gamma2)
            
            if loss_min > loss.item():
                loss_min = loss.item()
                best_S = S_pred_list
                best_I1 = I1_pred_list
                best_I2 = I2_pred_list
                best_R1 = R1_pred_list
                best_D1 = D1_pred_list
                best_R2 = R2_pred_list
                best_D2 = D2_pred_list

            dinn_iterator.set_description(f"epoch : {epoch+1}, loss:{loss.item():.4f}, alpha1:{float(self.alpha1.mean()):.4f}, alpha2:{self.alpha2[0]:.4f}, beta1:{self.beta1:.4f}, beta2:{self.beta2[0]:.4f}, gamma1:{self.gamma1:.4f}, gamma2:{self.gamma2[0]:.4f}")
        
        return S_pred_list, I1_pred_list, I2_pred_list, D1_pred_list, R1_pred_list, D2_pred_list, R2_pred_list, best_S, best_I1, best_I2, best_R1, best_D1, best_R2, best_D2, alpha_list, beta_list, gamma_list
```

```python
%%time
dinn = DINN(timesteps, cumulative_susceptible, cumulative_infected1, cumulative_infected2, 
            cumulative_dead1, cumulative_recovered1, cumulative_dead2, cumulative_recovered2, alpha1, device) #in the form of [t,S,I,D,R]

learning_rate = 1e-5
optimizer = optim.Adam(dinn.params, lr = learning_rate)
dinn.optimizer = optimizer

 
scheduler = torch.optim.lr_scheduler.CyclicLR(dinn.optimizer, base_lr=1e-5, max_lr=1e-3, step_size_up=1000, mode="exp_range", gamma=0.85, cycle_momentum=False)

dinn.scheduler = scheduler

try: 
    S_pred_list, I1_pred_list, I2_pred_list, D1_pred_list, R1_pred_list, D2_pred_list, R2_pred_list, best_S, best_I1, best_I2, best_R1, best_D1, best_R2, best_D2, alpha_list, beta_list, gamma_list = dinn.train(100000) #train
except EOFError:
    if dinn.save == 2:
        dinn.save = 3
        S_pred_list, I1_pred_list, I2_pred_list, D1_pred_list, R1_pred_list, D2_pred_list, R2_pred_list, best_S, best_I1, best_I2, best_R1, best_D1, best_R2, best_D2, alpha_list, beta_list, gamma_list = dinn.train(100000) #train
    elif dinn.save == 3:
        dinn.save = 2
        S_pred_list, I1_pred_list, I2_pred_list, D1_pred_list, R1_pred_list, D2_pred_list, R2_pred_list, best_S, best_I1, best_I2, best_R1, best_D1, best_R2, best_D2, alpha_list, beta_list, gamma_list = dinn.train(100000) #train
```
