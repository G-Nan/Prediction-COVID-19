```python
file = dic_files['서울']
```

```python
df = pd.DataFrame()

df['Date'] = file['stdDay']
df['Death_AC'] = file['deathCnt']               # 누적 사망자 수
df['Death'] = df['Death_AC'].diff().fillna(0)   # 일일 사망자
df['Inf_AC'] = file['defCnt']                   # 누적 감염자

Recover1 = 7     # 자가격리 해제 기간
Recover2 = 14    # 변이별 데이터가 없기 때문에 14일만 사용함

# 감염자(자가격리자) 수 계산
# t일 감염자 수 = t일 누적 감염자 수 - (t-14)일 누적 감염자 수
df['Infected'] = df['Inf_AC'] 
df.iloc[Recover2:, -1] = (df.iloc[Recover2:, -1]
                          .reset_index(drop = True)
                          .sub(df.iloc[:len(df) - Recover2, -1]))

Total = 9776000     # 서울시 인구 ( 2017년 기준 )
Reinfectable = 90   # 면역력 보유 기간

# 회복자(면역력 보유자) 수 계산
# t일 회복자 수 = (t-14)일 감염자 수 - (t-90)일 회복자 수 - t일 사망자 수
df['Recovered'] = 0
df.iloc[Recover2:, -1] = df.iloc[:len(df) - Recover2, 3]
df['Recovered_Ori'] = df['Recovered']
df.iloc[Reinfectable:, -1] = (df.iloc[Reinfectable:, -1]
                              .reset_index(drop = True)
                              .sub(df.iloc[:len(df) - Reinfectable, -1]))
df['Recovered'] = df['Recovered'] - df['Death']

# 생존자 수 계산
# t일 생존자 수 = t일 총 인구 - t일 감염자 - t일 회복자 - t일 사망자
df['Survived'] = Total
df['Survived'] = df['Survived'] - df['Infected'] - df['Recovered'] - df['Death']


df[500:550]
```

```python
df['alpha'] = (-1 * Total * (df['Survived'].diff()[1:].reset_index(drop = True))) / (df['Survived'] * df['Infected'])
df['beta'] = (df['Recovered'].diff()[1:].reset_index(drop = True)) / df['Infected']
df['gamma'] = (df['Death'].diff()[1:].reset_index(drop = True)) / df['Infected']
df.head(20)
```
