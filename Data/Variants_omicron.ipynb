```python
df_variants = pd.read_csv('Data/Variants/Variants.csv')
df_variants = df_variants.fillna(0)
df_variants
```

```python
path = 'Data/Cleaned_deathCnt/*.csv'
dic_files = Load_files.load_files(path, -14, -12)
data = dic_files['서울']
data
```

```python
path = 'Data/SIR_Origin/*.csv'
dic_SIRs = Load_files.load_files(path, -17, -15)
data = dic_SIRs['서울']
data
```

```python
omicron_22D = df_variants['22D'][868:-3].reset_index(drop = True)
omicron_23A = df_variants['23A'][868:-3].reset_index(drop = True)
dic_omicron = {}

for city, data in dic_files.items():
    data['dailyDeath'] = data['deathCnt'].diff().fillna(0)
    df_omicron = data.iloc[1064:1064+137, [4, 10, 11]].reset_index(drop = True).copy()
    df_omicron['22D_Infected'] = df_omicron['incDec']*omicron_22D
    df_omicron['23A_Infected'] = df_omicron['incDec']*omicron_23A
    df_omicron['22D_dead'] = 0
    df_omicron.iloc[7:, -1] = omicron_22D[:-7]*df_omicron['dailyDeath'][14:].reset_index(drop = True)
    df_omicron['23A_dead'] = 0
    df_omicron.iloc[7:, -1] = omicron_23A[:-7]*df_omicron['dailyDeath'][14:].reset_index(drop = True)
    df_omicron = df_omicron[['stdDay', 'incDec','22D_Infected', '23A_Infected', '22D_dead', '23A_dead']]
    
    df_omicron.iloc[:, 2:] = df_omicron.iloc[:, 2:].expanding().sum() 
    df_omicron['Susceptible'] = dic_SIRs[city]['Susceptible'][1064:1064+137].reset_index(drop = True)

    df_omicron['22D_Recovered'] = 0
    df_omicron.iloc[7:, -1] = df_omicron.iloc[:len(df_omicron)-7, 2]
    df_omicron['23A_Recovered'] = 0
    df_omicron.iloc[7:, -1] = df_omicron.iloc[:len(df_omicron)-7, 3]

    df_omicron.iloc[7:, 2] = (df_omicron.iloc[7:, 2]
                             .reset_index(drop = True)
                             .sub(df_omicron.iloc[:len(df_omicron)-7, 2]))
    
    df_omicron.iloc[7:, 3] = (df_omicron.iloc[7:, 3]
                             .reset_index(drop = True)
                             .sub(df_omicron.iloc[:len(df_omicron)-7, 3]))

    df_omicron['22D_alpha'] = (((df_omicron['22D_Infected'].shift(-1) - df_omicron['22D_Infected']) 
                               + (df_omicron['22D_Recovered'].shift(-1) - df_omicron['22D_Recovered'])
                               + (df_omicron['22D_dead'].shift(-1) - df_omicron['22D_dead']))
                                    * (dic_SIRs[city]['Susceptible'][0] / (df_omicron['Susceptible'] * df_omicron['22D_Infected'])))
    
    df_omicron['23A_alpha'] = (((df_omicron['23A_Infected'].shift(-1) - df_omicron['23A_Infected']) 
                               + (df_omicron['23A_Recovered'].shift(-1) - df_omicron['23A_Recovered'])
                               + (df_omicron['23A_dead'].shift(-1) - df_omicron['23A_dead']))
                                    * (dic_SIRs[city]['Susceptible'][0] / (df_omicron['Susceptible'] * df_omicron['23A_Infected'])))

    df_omicron['22D_beta'] = (df_omicron['22D_Recovered'].shift(-1) - df_omicron['22D_Recovered']) / df_omicron['22D_Infected']
    df_omicron['23A_beta'] = (df_omicron['23A_Recovered'].shift(-1) - df_omicron['23A_Recovered']) / df_omicron['23A_Infected']
    
    df_omicron['22D_gamma'] = (df_omicron['22D_dead'].shift(-1) - df_omicron['22D_dead']) / df_omicron['22D_Infected']
    df_omicron['23A_gamma'] = (df_omicron['23A_dead'].shift(-1) - df_omicron['23A_dead']) / df_omicron['23A_Infected']    
    


    
    df_omicron = df_omicron[['stdDay', 'Susceptible', 
                             '22D_Infected', '22D_Recovered', '22D_dead', '22D_alpha', '22D_beta', '22D_gamma',
                             '23A_Infected', '23A_Recovered', '23A_dead', '23A_alpha', '23A_beta', '23A_gamma']]
    
    df_omicron = df_omicron.fillna(0)
    
    dic_omicron[city] = df_omicron
```

```python
for city, data in dic_omicron.items():
    data.to_csv(f"Data/Variants/omicron/{city}_omicron.csv", encoding = 'cp949', index = False)
```
