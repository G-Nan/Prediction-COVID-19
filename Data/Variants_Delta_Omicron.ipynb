```python
delta = df_variants['Delta'][252:-4].reset_index(drop = True)
omicron = df_variants['Omicron'][252:-4].reset_index(drop = True)

for city, data in dic_files.items():
    df_delta_omicron = dic_files[city].iloc[448:-19, [4, 10]].reset_index(drop = True).copy()
    df_delta_omicron['delta_variants'] = df_delta_omicron['incDec']*delta
    df_delta_omicron['omicron_variants'] = df_delta_omicron['incDec']*omicron
    df_delta_omicron.to_csv(f"Data/Variants/Infected/delta_omicron/{city}_delta_omicron.csv", encoding = 'cp949', index = False)
```
