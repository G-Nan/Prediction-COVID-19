```python
import pandas as pd
import numpy as np
import glob
```

```python
file_list = glob.glob('./Prediction-COVID-19/Data/CleanedData/*')
file_list_csv= [file for file in file_list if file.endswith('.csv')]

total_df = pd.DataFrame()
for file in file_list_csv:
    df = pd.read_csv(file, usecols=['deathCnt', 'defCnt', 'gubun', 'incDec', 'isolClearCnt', 'localOccCnt', 'overFlowCnt', 'stdDay'])
    total_df = pd.concat([total_df, df])

total_df.rename(columns={'deathCnt':'Acc_D', 'defCnt':'Acc_I', 'gubun':'region', 'incDec':'I', 'isolClearCnt':'Acc_R', 'styDay':'date'}, inplace=True)
total_df
```

```python
pop_df = pd.read_csv('./data/202304인구수.csv', usecols=['행정구역', '2023년04월_총인구수'], encoding='cp949')
pop_df['region'] = [region[:2] for region in pop_df['행정구역'].values]
pop_df.drop('행정구역', axis=1, inplace=True)
pop_df.rename(columns={'2023년04월_총인구수':'popN'}, inplace=True)
pop_df['popN'].replace(',','', regex=True, inplace=True)
pop_df['popN'] = pop_df['popN'].astype(float)
pop_df.info()
```

```python
df = pd.merge(total_df, pop_df, how='inner', on='region')
df_N = df.set_index(['region', 'stdDay']).sort_index()
df_N.info()
```

```python
regions = df_N.index.map(lambda x:x[0]).unique()

dic_S = {}
df_S = pd.DataFrame()

for region in regions:
    region_df = df_N.loc[region]
    
    for i in range(len(region_df)):
        if i == 0:
            region_df['S'] = region_df['popN'][0]
        else:
            region_df['S'][i] = region_df['S'][i-1] - region_df['I'][i]
            
    dic_S[region] = region_df
    
    region_df['region'] = region
    region_df.reset_index(inplace=True)
    df_S = pd.concat([df_S, region_df])
    
```

```python
dic_D = {}
df_D = pd.DataFrame()

for region in regions:
    df = dic_S[region]
    
    for i in range(len(df)):
        if i == 0:
            df['D'] = df['Acc_D'][0]
        else:
            df['D'][i] = df['Acc_D'].diff()[i]
            
    dic_D[region] = df
    df.set_index('index', inplace=True)
    df_D = pd.concat([df_D, df])
```
